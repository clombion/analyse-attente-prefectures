## I. Import du du fichier

```python
#import des librairies de manipulation de données
from pandas import *
import numpy as np
from scipy import stats, integrate

#import des librairies de visualisation
import seaborn as sns; sns.set(style="ticks", color_codes=True)
import matplotlib as mpl;
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#on charge le csv en indiquant que le séparateur est la virgule et que la première ligne contient les en-têtes
df = read_csv('CIMADE_cleaned_v1.csv', sep=',', header=0)
```

On a ici créé un dataframe appelé "data". Un dataframe est une "vue" du fichier CSV, c'est à dire un tableau virtuel qu'on pourra modifier à volonté sans toucher au fichier initial.

Plusieurs paramètres ont dû être ajoutés pour éviter des erreurs de lecture du CSV:
* sep=";" indique à Pandas que le fichier CSV est délimité par des points-virgules (les fichiers créés par des programmes anglophones utilisent des virgules, contrairement à la convention française du point virgule)
* header=0 nous permet d'indiquer que la première ligne contient les en-têtes du CSV (la première valeur d'une liste pour Python est le numéro 0, et la seconde valeur est le numéro 1)

## II. Préparation des données
Open Refine a permis d'avoir une bonne idée du fichier. On va cependant jeter un nouveau coup d'oeil au fichier pour s'assurer que tout est place pour commencer l'analyse avec Pandas.

```python
#on affiche un résumé synthétiques des métadonnées du dataframe: nombre de lignes, types de données etc
df.info()

#on affiche 5 lignes choisies au hasard dans le dataframe
df.sample(5)
```

On voit ici que notre fichier fait 234872 lignes et 7 colonnes. Il est stocké en mémoire (RAM), en prend 12,5MB. Pandas peut travailler avec des fichiers bien plus gros, mais ce n'est pas dit que votre ordi puisse suivre !

Un problème apparaît : comme OpenRefine avant lui, Pandas ne reconnaît pas les dates en tant que telles. Le numéro du département est bien reconnu comme un chiffre entier (int64), mais les éléments des autres colonnes sont tous considérés comme du text (object). Cela rend impossible de, par exemple, trouver la date la plus récente de la colonne "date_sondage." Il va faloir donc convertir ces colonnes avant de pouvoir travailler avec.

Une autre chose à savoir est qu'il n'est pas très utile d'avoir les heures et les dates séparées dans Pandas. Pandas ne reconnait pas les heures en tant que telles, à moins qu'elles soient associérs avec une date. Il vaut mieux donc regrouper tout ça dans une même colonne.

```python
#on nettoie les données de date_sondage et on la fusionne avec heure_sondage
df['date_sondage'] = df['date_sondage'].str.replace('T00:00:00Z',' ') + df['heure_sondage']
df = df.drop('heure_sondage',1)

#on affiche les 5 premières lignes du nouveau dataframe
df.head(5)
```

On peut maintenant changer le type des trois colonnes qui indiquent des dates

```python
#on utilise to.datetime pour assigner un type adapté aux données temporelle
df['date_sondage'] = to_datetime(df['date_sondage'])
df['proposition1'] = to_datetime(df['proposition1'], errors='coerce')
df['proposition2'] = to_datetime(df['proposition2'], errors='coerce')

#on vérifie les différents types du dataframe
df.dtypes
```

Et tant qu'à faire, on va rajouter les infos de delais directement.

```python
#on calcule le delai entre la date de scraping et les deux propositions.
df['delai'] = df['proposition1'] - df['date_sondage']
df['delai2'] = df['proposition2'] - df['date_sondage']

#infos de délai en jour, mais dans un format float qui facilitera certaines analyses ultérieures
df['delai_jour'] = df['delai'].dt.total_seconds() / (24 * 60 * 60)

df['delai_jour'].max()
```

Dans beaucoup de jeux de données, on ignore en général les valeurs absentes (NaN). Cependant, ici l'absence de valeur dans les colonnes liées aux rendez-vous est importante à repérer vu qu'elle indique une impossibilité de prendre un rendez-vous. On va donc bien l'identifier (et préparer de futures analyses) en créant une dernière colonne booléenne basée sur la présence ou l'absence de NaN. 

```python
#on crée une colonne "impossible" en comptant les NaN (isnull) de la colonne délai puis en les transformant en 1 ou 0
df['impossible'] = df['delai'].isnull().astype(int)
df.sample(5)
```

Une dernière chose à faire est de changer l'index (le compte des colonnes, tout à gauche). L'index permet de se repérer dans la tableau, et permet de récupérer une ligne précise par son numéro. Mais on a aussi une autre colonne qui a des valeurs uniques pour chaque ligne: date_sondage. Et de façon générale, lorsque l'on travaille avec une série temporelle, il vaut mieux mettre le temps en index de façon à pouvoir faire des opérations sur le temps facilement.

```python
#on assigne date_sondage à l'index
df.index = df['date_sondage']

df.head(5)
```
